```python
indian['cook_time'].plot.hist(bins = 20)
plt.legend()
```

```python
indian['cook_time'].plot.hist(bins = 10)
plt.legend()
#it does not show the distribution according to lower minutes, just 3 bins are filled
```

```python
indian['cook_time'].plot.hist(bins = 50, figsize = (20,9))
plt.xticks(np.arange(0, indian['cook_time'].max(),20))
```

```python
indian['prep_time'].plot.hist(alpha = 0.5, bins = 50, figsize = (20,9))
indian['cook_time'].plot.hist(alpha = 0.5, bins = 50, figsize = (20,9))
plt.xticks(np.arange(0, indian['prep_time'].max(),20))
plt.xticks(rotation = 45)
```

```python
meantime = indian.groupby('prep_time').mean()
meantime
```

```python
meantime.plot.area()
```

```python
#pie chart
#compare one or more categorial variables

indian['flavor_profile'].value_counts().plot.pie(figsize = (10,6))
plt.legend()
```

```python
colors = ['orchid', 'mediumseagreen', 'steelblue', 'goldenrod', 'teal']
indian['flavor_profile'].value_counts().plot.pie(figsize = (10,6), colors = colors, shadow = True, title = 'Flavor Profile')
plt.legend()
```
